### Import Libraries

```python
import numpy
import cv2 as cv
import matplotlib.pyplot as plt
from PIL import Image
from IPython.display import Image
%matplotlib inline
```

### Import Image

```python
image = cv.imread("Images\Apollo_11_Launch.jpg", 1)

plt.imshow(image[:,:,::-1])
```

### Draw a line on image

```python
line_image = image.copy()

cv.line(line_image,(200,100), (450,200), (0,255,0), thickness=9, lineType=cv.LINE_AA);

plt.imshow(line_image[:,:,::-1])
```

### Draw a circle

```python
circle_image = image.copy()

cv.circle(circle_image, (600,400), 100, (0,255,255), thickness= 9, lineType=cv.LINE_8)

plt.imshow(circle_image[:,:,::-1])
```

### Draw a Rectangle

```python
rect_image = image.copy()

cv.rectangle(rect_image, (500,100), (700,600), (0,255,255), thickness=5, lineType=cv.LINE_AA);

plt.imshow(rect_image[:,:,::-1])
```

### Adding Text

```python
text_image = image.copy()

text = "This is an Appolo launch event Image"

cv.putText(text_image, text, (200,680), fontFace=cv.FONT_HERSHEY_COMPLEX, fontScale=1.5, color=(0,255,255), thickness=5, lineType=cv.LINE_AA);

plt.imshow(text_image[:,:,::-1])
```
