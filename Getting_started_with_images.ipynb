### Split and Merge

```python
#loading the Scenery image and splitting it into B,G,R
img_scene = cv.imread('/content/scene.jpg', 1)
b,g,r = cv.split(img_scene)

#Showing the channels
plt.figure(figsize=[20,5])
plt.subplot(141); plt.imshow(r, cmap='gray'); plt.title("Red Channel");
plt.subplot(142); plt.imshow(g, cmap='gray'); plt.title("Green Channel");
plt.subplot(143); plt.imshow(b, cmap='gray'); plt.title("Blue Channel");

#Merging the channels
imgMerged = cv.merge((b,g,r))
plt.subplot(144); plt.imshow(imgMerged[:,:,::-1]); plt.title("Merged Image")
```

```python
img_scene_rgb = cv.cvtColor(img_scene, cv.COLOR_BGR2RGB)
plt.imshow(img_scene_rgb)
```

### Converting from BGR to HSV color space

```python
# Split image imto h,s,v
img_scene_hsv = cv.cvtColor(img_scene, cv.COLOR_BGR2HSV)
h,s,v = cv.split(img_scene_hsv)

#show the h,s,v components
plt.figure(figsize=[20,5])
plt.subplot(151); plt.imshow(h, cmap='gray'); plt.title("H Channel");
plt.subplot(152); plt.imshow(s, cmap='gray'); plt.title("S Channel");
plt.subplot(153); plt.imshow(v, cmap='gray'); plt.title("V Channel");
plt.subplot(154); plt.imshow(img_scene_hsv[:,:,::-1]); plt.title("HSV Image")
plt.subplot(155); plt.imshow(img_scene_rgb); plt.title("Original Image")
```

### Modifying individual Channels

```python
'''
1. Create a new h value.
2. Merge the new h, s, v values.
3. Convert from HSV to RGB.
'''
h_new = h+10
imgMerged_new = cv.merge((h_new,s,v))
img_new = cv.cvtColor(imgMerged_new, cv.COLOR_HSV2RGB)

#Show the channels
plt.figure(figsize=[20,5])
plt.subplot(141); plt.imshow(h_new, cmap='gray'); plt.title("New h channel");
plt.subplot(142); plt.imshow(s, cmap='gray'); plt.title("s channel");
plt.subplot(143); plt.imshow(v, cmap='gray'); plt.title("v channel");
plt.subplot(144); plt.imshow(img_new); plt.title("Modified Image");
```

### Saving an image

```python
cv.imwrite("New_modified_image.png", img_new)

Image(filename='New_modified_image.png')
```
